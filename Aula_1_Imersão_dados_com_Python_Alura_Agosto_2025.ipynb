````python
import pandas as pd
````

````python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv") ##Ler a biblioteca pandas - Arquivo utilizado em aula - Ler uma base de dados usando Pandas - Arquivo CSV
````

````python
df.columns ##nome das colunas detalhadas
````

````python
df.columns = ['Ano', 'Senioridade', 'Contrato', 'Cargo',
              'Salario', 'Moeda', 'USD', 'Residencia',
              'Remoto', 'Empresa', 'Tamanho_Empresa']
display(df.head(10)) ##Codigo usado por IA para fazer alteração do nome das colunas
````

````python
df['Senioridade'].value_counts() ##Contar os valores de uma coluna - SE - Senior - MI - intermediate - EN - Junior - EX - Executive
````

````python
df['Contrato'].value_counts()
````

````python
df.Remoto.value_counts()
````

````python
df.Tamanho_Empresa.value_counts()
````

````python
substituir_senioridade = {
    'SE': 'Senior',
    'MI': 'Intermediate',
    'EN': 'Junior',
    'EX': 'Executive'
}
df['Senioridade'] = df['Senioridade'].map(substituir_senioridade)
df['Senioridade'].value_counts()
````

````python
substituir_contrato = {
    'PT': 'Part-Time',
    'FL': 'Freelance',
    'CT': 'Contrato',
    'FT': 'Full-Time'
}
df['Contrato'] = df['Contrato'].map(substituir_contrato)
df['Contrato'].value_counts()
````

````python
substituir_Tamanho_Empresa = {
    'S': 'Startup',
    'M': 'Mid-sized',
    'L': 'Large'
}
df['Tamanho_Empresa'] = df['Tamanho_Empresa'].map(substituir_Tamanho_Empresa)
df['Tamanho_Empresa'].value_counts()
````

````python
substituir_remoto = {
    100: 'Remoto',
    50: 'Hibrido',
    0: 'Presencial'
}
df['Remoto'] = df['Remoto'].map(substituir_remoto)
df['Remoto'].value_counts()
````
