````python
import numpy as np ##np.nan esta sendo utilizado nesse situação para colocar valores nulos para simular salarios nulos.
##Codigo abaixo esta fazendo uma simulação de uma nova dataframe e criação para fazer uma analise de salarios medianos e media para exemplificar o que acontece quando temos valores nulos em nossos dados.
#df_salarios é o nome da nossa base de dados

df_salarios = pd.DataFrame({'Nome': ['João', 'Clara', 'Carlos', 'Danielle', 'Val'],
                            'Salario': [4500, np.nan, 5200, np.nan, 100000]
})
#Fazer o calculo da Média

df_salarios['salario_media'] = df_salarios['Salario'].fillna(df_salarios['Salario'].mean().round(2)) ##fillna é utilizado para fazer o preenchimento dos valores nulos que vai ser calculado a partir da média.
##round(2) é feito para arredondar valores calculados
df_salarios
````

````python
##Fazer calculo da Mediana

df_salarios['salario_mediana'] = df_salarios['Salario'].fillna(df_salarios['Salario'].median()) ##Analise feita para valores medianos lembrando de fazer a base de dados salario_mediana. Tambem foi retirado valores round para arredondamento
df_salarios
````

````python
df_temperaturas = pd.DataFrame({ 'Dia': ['Segundo', 'Terça', 'Quarta', 'Quinta', 'Sexta'],
                                 'Temperatura': [30, np.nan, 31, np.nan, 33]
})
df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill() ##criar nova coluna
df_temperaturas

##ffill - completa com valores anteriores. Nesse caso nos meus dados foi completado com valores anteriores existentes
````

````python
##Grafico interativo
##Usando uma biblioteca em Python - Plotly


import plotly.express as px
````

````python
# Calculate the average salary by seniority
senioridade_media_salario = df_limpo.groupby('Senioridade')['USD'].mean().sort_values(ascending=False).reset_index()

# Define the order of seniority levels
ordem_senioridade = ['Junior', 'Intermediate', 'Senior', 'Executive']

# Create the bar plot using Plotly Express
fig = px.bar(senioridade_media_salario,
             x='Senioridade',
             y='USD',
             title='Média Salarial por Senioridade',
             labels={'Senioridade': 'Nível de Senioridade', 'USD': 'Média Salarial Anual (USD)'},
             category_orders={'Senioridade': ordem_senioridade})

# Show the plot
fig.show()
````

````python
remoto_contagem = df_limpo['Remoto'].value_counts().reset_index()
remoto_contagem.columns = ['Tipo_Trabalho', 'Quantidade']

fig = px.pie(remoto_contagem,
             names='Tipo_Trabalho',
             values='Quantidade',
             title='Proporção dos Tipos de Trabalho',
             hole=0.5
)

fig.update_traces(textinfo='percent+label')

fig.show()
````

````python
import pycountry
````

````python
##Função para Converter ISO-2 para ISO-3

def iso2_to_iso3(code):
    try:
        return pycountry.countries.get(alpha_2=code).alpha_3
    except:
        return None

##Criar nova coluna com código ISO-3
df_limpo['residencia_iso3'] = df_limpo['Residencia'].apply(iso2_to_iso3)

##Calcular médial salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['Cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['USD'].mean().reset_index()

##Gerar o Mapa

fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='USD',
                    color_continuous_scale='rdylgn',
                    title='Média Salarial de Data Scientist por País',
                    labels={'USD': 'Média Salarial Anual (USD)', 'residencia_iso3': 'País'})


fig.show()
````
